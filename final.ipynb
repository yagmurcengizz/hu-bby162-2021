```python
#Kütüphane Kataloğu Otomasyonu
#Veri Tabanı: /content/drive/MyDrive/HU BBY162/Colab Notebooks/final.txt

dosya = "/content/drive/MyDrive/HU BBY162/Colab Notebooks/final.txt"

def eserleriListele():                    
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()


def eserKaydet():
  eserAdi = input("Eser Adı: ")
  yazarAdi = input("Yazar Adı: ")
  yayinevi = input("Yayınevi: ")
  basimTarihi = input("Basım Tarihi: ")
  isbnNumarasi = input("ISBN Numarası: ")
  f = open(dosya, 'a')
  f.write(eserAdi + "," + yazarAdi + "," + yayinevi + "," + basimTarihi + "," + isbnNumarasi + '\n' )
  print("Yeni eser başarıyla kayıt edilmiştir!")
  f.close


def cikisYap():
  f = open(dosya, 'r')
  print("Çıkış yapılıyor...")
  f.close


secimMenusu = 0
while not secimMenusu == 4 :
    print("Kütüphane Kataloğu Otomasyonuna Hoşgeldiniz!")
    print("1)Eser Kaydet")
    print("2)Eserleri Listele")
    print("3)Çıkış Yap ")
    secimMenusu = int (input("Lütfen seçmek istediğiniz işlemi rakam olarak giriniz:")) 
                             
    
    if secimMenusu == 1 :
        eserKaydet()
    elif secimMenusu == 2 :
        eserleriListele()
    elif secimMenusu == 3 :  
        print("Çıkış yapılıyor...")
        break
    else:
      print("Lütfen seçmek istediğiniz işlemi rakam olarak giriniz:")
      continue
```
